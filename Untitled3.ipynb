1. LOAD THE DATASETS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score


try:
    temp_anomaly = pd.read_csv('temperature-anomaly.csv')
    temp_rise_total = pd.read_csv('contribution-temp-rise-degrees.csv')
    temp_rise_gas = pd.read_csv('contribution-to-temp-rise-by-gas.csv')
    ghg_region = pd.read_excel('ghg-emissions-by-world-region.xlsx')
    fossil_warming = pd.read_excel('global-warming-fossil.xlsx')
    land_warming = pd.read_excel('global-warming-land.xlsx')
    print("All datasets loaded successfully!")
except FileNotFoundError as e:
    print(f"Error loading files: {e}. Check file paths.")


```

2. DATA CLEANING & MATCHING THE GLOBAL LEVEL

```python


global_temp = temp_anomaly[temp_anomaly['Entity'] == 'Global'].copy()

global_temp.rename(columns={
    'Global average temperature anomaly relative to 1961-1990': 'temp_anomaly'
}, inplace=True)


def filter_global(df):

    return df[df['Entity'].isin(['World', 'Global'])].copy()

df_total = filter_global(temp_rise_total)
df_gas = filter_global(temp_rise_gas)
df_ghg = filter_global(ghg_region)
df_fossil = filter_global(fossil_warming)
df_land = filter_global(land_warming)


df_total.rename(columns={'Change in global mean surface temperature caused by greenhouse gas emissions': 'total_ghg_warming'}, inplace=True)
df_gas.rename(columns={
    'Change in global mean surface temperature caused by nitrous oxide emissions': 'n2o_warming',
    'Change in global mean surface temperature caused by methane emissions': 'chg_methane_warming',
    'Change in global mean surface temperature caused by CO₂ emissions': 'co2_warming'
}, inplace=True)
df_ghg.rename(columns={'Annual greenhouse gas emissions in CO₂ equivalents': 'raw_ghg_emissions'}, inplace=True)
df_fossil.rename(columns={'Change in global mean surface temperature caused by greenhouse gas emissions from fossil fuels and industry': 'fossil_warming'}, inplace=True)
df_land.rename(columns={'Change in global mean surface temperature caused by greenhouse gas emissions from agriculture and land use': 'land_warming'}, inplace=True)
```

```python
print(df_gas.columns)
```

3. THE MASTER MERGE (FIXED)

```python

master_df = global_temp[['Year', 'temp_anomaly']]


datasets = [df_total, df_gas, df_ghg, df_fossil, df_land]
target_columns = [
    'total_ghg_warming',
    ['n2o_warming', 'chg_methane_warming', 'co2_warming'],
    'raw_ghg_emissions',
    'fossil_warming',
    'land_warming'
]
for df, cols in zip(datasets, target_columns):

    if isinstance(cols, list):
        master_df = pd.merge(master_df, df[['Year'] + cols], on='Year', how='inner')
    else:
        master_df = pd.merge(master_df, df[['Year', cols]], on='Year', how='inner')


master_df.dropna(inplace=True)

print(f"Master dataset consolidated. Shape: {master_df.shape}")
print(master_df.head())
```

4. EXPLORATORY DATA ANALYSIS PLOTS

```python

sns.set_theme(style="whitegrid")


plt.figure(figsize=(10, 5))
plt.plot(master_df['Year'], master_df['temp_anomaly'], color='crimson', label='Temperature Anomaly', linewidth=2)
plt.axvline(x=1980, color='gray', linestyle='--', label='1980 Base Split Point')
plt.title('Acceleration of Global Temperature Anomalies Over Time')
plt.xlabel('Year')
plt.ylabel('Anomaly (Relative to 1961-1990 Baseline)')
plt.legend()
plt.tight_layout()
plt.savefig('eda_temperature_anomaly_trend.png')
plt.show()


plt.figure(figsize=(10, 5))
plt.plot(master_df['Year'], master_df['co2_warming'], label='CO2 Warming Contribution', color='darkblue')
plt.plot(master_df['Year'], master_df['chg_methane_warming'], label='Methane Warming Contribution', color='orange')
plt.plot(master_df['Year'], master_df['n2o_warming'], label='Nitrous Oxide Contribution', color='green')
plt.title('Warming Contribution Trajectories by Individual Gas Type')
plt.xlabel('Year')
plt.ylabel('Temperature Rise Contribution (°C)')
plt.legend()
plt.tight_layout()
plt.savefig('eda_gas_contributions.png')
plt.show()
```

 5. CORRELATION MATRIX & FEATURE FILTERING

```python

plt.figure(figsize=(8, 6))

corr_matrix = master_df.drop(columns=['Year']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Analysis of Global Warming Factors')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()


selected_features = ['co2_warming', 'land_warming']
print("Features isolated to eliminate multicollinearity issues:", selected_features)
```

6. TEMPORAL DATA SPLIT

```python

train_mask = (master_df['Year'] >= 1980) & (master_df['Year'] <= 1990)
test_mask = (master_df['Year'] >= 1991)

X_train = master_df.loc[train_mask, selected_features]
y_train = master_df.loc[train_mask, 'temp_anomaly']

X_test = master_df.loc[test_mask, selected_features]
y_test = master_df.loc[test_mask, 'temp_anomaly']

print(f"Training samples (1980-1990): {len(X_train)}")
print(f"Testing samples (1991 onwards): {len(X_test)}")

```

7. MODEL TRAINING & PERFORMANCE TESTING

```python
model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n--- MODEL PERFORMANCE METRICS ---")
print(f"Model Coefficients: {dict(zip(selected_features, model.coef_))}")
print(f"Model Intercept: {model.intercept_:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}°C")
print(f"R-squared Score (R²): {r2:.4f}")
```

8. FUTURE 10-YEAR RISK PROJECTION

```python

last_recorded_year = master_df['Year'].max()
print(f"\nExtrapolating from last recorded dataset year: {last_recorded_year}")


recent_trends = master_df[master_df['Year'] >= (last_recorded_year - 10)]
co2_annual_growth = recent_trends['co2_warming'].diff().mean()
land_annual_growth = recent_trends['land_warming'].diff().mean()

future_years = list(range(int(last_recorded_year + 1), int(last_recorded_year + 11)))
future_co2 = [master_df['co2_warming'].iloc[-1] + (i * co2_annual_growth) for i in range(1, 11)]
future_land = [master_df['land_warming'].iloc[-1] + (i * land_annual_growth) for i in range(1, 11)]

future_scenario = pd.DataFrame({
    'Year': future_years,
    'co2_warming': future_co2,
    'land_warming': future_land
})


future_scenario['projected_anomaly'] = model.predict(future_scenario[selected_features])

print("\n--- AETHELGARD RE: 10-YEAR FORECAST FOR RISK PROFILING ---")
print(future_scenario[['Year', 'projected_anomaly']])


plt.figure(figsize=(10, 5))
plt.plot(master_df['Year'], master_df['temp_anomaly'], label='Historical Observed Data', color='black')
plt.plot(future_scenario['Year'], future_scenario['projected_anomaly'], label='10-Year Forward Model Projection', color='blue', linestyle='--')
plt.title('Aethelgard Re: Long-Term Temperature Anomaly Projections')
plt.xlabel('Year')
plt.ylabel('Anomaly Change (°C)')
plt.legend()
plt.tight_layout()
plt.savefig('actuarial_10_year_forecast.png')
plt.show()
```
